```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
def inception_module(input):
    path1 = layers.Conv2D(64, (1, 1), padding='same', activation='relu')(input)
    path2 = layers.Conv2D(128, (3, 3), padding='same', activation='relu')(input)
    path3 = layers.Conv2D(32, (5, 5), padding='same', activation='relu')(input)
    path4 = layers.MaxPooling2D((3, 3), strides=(1, 1), padding='same')(input)
    output = layers.concatenate([path1, path2, path3, path4], axis=-1)

    return output

input_shape = (28, 28, 192)
input_tensor = layers.Input(shape=input_shape)
output_tensor = inception_module(input_tensor)

# Create the model
model = models.Model(inputs=input_tensor, outputs=output_tensor)
model.summary()
```
